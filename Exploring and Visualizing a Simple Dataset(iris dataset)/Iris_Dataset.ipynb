## **Load** **data**
Let's load the iris flower dataset using scikit-learn's build-in datasets

```python
# To load iris dataset
df = sns.load_dataset('iris')
```

## **Structure** **Check**

```python
# Shape:
df.shape
```

```python
# Columns:
df.columns
```

```python
# getting first 5 rows
df.head()
```

## **Scatter** **Plot**

```python
# Scatter plot between sepal length and sepal width
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.show()
```

## **Histogram**

```python
# Distribution of sepal width
sns.histplot(df['sepal_width'], kde=True)
plt.show()
```

## **Box** **Plot**

```python
# To show spread and outliers of sepal width according to species
sns.boxplot(x='species', y='sepal_width', data=df)
plt.show()
```

## **Conclusion**

Based on the Exploratory Data Analysis (EDA) of the Iris dataset, I have found the following:

**Data** **Structure**: The dataset contains 150 rows and 5 columns (sepal_length, sepal_width, petal_length, petal_width, and species).


**Species** **Separation**: The scatter plot clearly shows that the Setosa species is easily distinguishable from Versicolor and Virginica due to its smaller petal length and width.


**Distribution**: Roughly normal distribution, with most values centered around 3.0 cm.


**Outliers**: There are a few outliers in the sepal width of the Virginica species.
